```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Create synthetic data
np.random.seed(42)
X = np.random.rand(50, 2) * [5, 4]

# Step 2: Define initial centroids manually
initial_centroids = np.array([[2.5, 1.8], [4.0, 2.6]])

# Step 3: Perform 1 iteration of KMeans using initial seeds
kmeans = KMeans(n_clusters=2, init=initial_centroids, n_init=1, max_iter=1, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
new_centroids = kmeans.cluster_centers_

# Step 4: Plotting
plt.figure(figsize=(5, 4))

# Plot cluster 1 points with '+'
plt.scatter(X[labels == 0][:, 0], X[labels == 0][:, 1], marker='.', color='blue', label='Cluster 1')

# Plot cluster 2 points with 'o'
plt.scatter(X[labels == 1][:, 0], X[labels == 1][:, 1], facecolors='none', edgecolors='green', label='Cluster 2')

# Plot initial centroids as big 'X'
plt.scatter(initial_centroids[:, 0], initial_centroids[:, 1], color='black', marker='x', s=200, linewidths=2.5, label='Initial Centroids')

# Plot new centroids as '*'
plt.scatter(new_centroids[:, 0], new_centroids[:, 1], color='black', marker='*', s=150, label='New Centroids')

# Draw arrows from initial to new centroids
for i in range(2):
    plt.arrow(initial_centroids[i, 0], initial_centroids[i, 1],
              new_centroids[i, 0] - initial_centroids[i, 0],
              new_centroids[i, 1] - initial_centroids[i, 1],
              color='black', width=0.01, head_width=0.15, length_includes_head=True)

plt.xlim(0, 6)
plt.ylim(0, 4)
plt.xticks(np.arange(0, 7, 1))
plt.yticks(np.arange(0, 5, 1))
plt.title("Re-computation of Centroid", fontsize=10)
plt.grid(False)

# Remove top and right spines for a clean look
for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Step 1: Generate synthetic data
np.random.seed(42)
X = np.random.rand(50, 2) * [5.5, 4]  # Rescaled for range similar to plot

# Step 2: Run full K-Means until convergence
kmeans = KMeans(n_clusters=2, init='k-means++', n_init=10, random_state=42)
kmeans.fit(X)
labels = kmeans.labels_
centroids = kmeans.cluster_centers_

# Step 3: Plotting
plt.figure(figsize=(5, 4))

# Plot cluster points
plt.scatter(X[labels == 0][:, 0], X[labels == 0][:, 1], marker='.', color='blue', label='Cluster 1')
plt.scatter(X[labels == 1][:, 0], X[labels == 1][:, 1], facecolors='none', edgecolors='green', label='Cluster 2')

# Plot final centroids
plt.scatter(centroids[:, 0], centroids[:, 1], color='black', marker='x', s=200, linewidths=2.5, label='Centroids')

# Axes setup
plt.xlim(0, 6)
plt.ylim(0, 4)
plt.xticks(np.arange(0, 7, 1))
plt.yticks(np.arange(0, 5, 1))
plt.title("Centroids after Convergence", fontsize=10)
plt.grid(False)

# Clean borders like in the reference image
for spine in ['top', 'right']:
    plt.gca().spines[spine].set_visible(False)

plt.tight_layout()
plt.show()
```

```python
from google.colab import files
import numpy as np
import cv2
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# Upload and load image
uploaded = files.upload()
filename = next(iter(uploaded))
image = cv2.imread(filename)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Flatten image pixels
pixel_vals = image.reshape((-1, 3))
pixel_vals = np.float32(pixel_vals)

# === UNDERFITTING: K = 2 ===
K_underfit = 2
kmeans_underfit = KMeans(n_clusters=K_underfit, random_state=42)
kmeans_underfit.fit(pixel_vals)
centers_underfit = np.uint8(kmeans_underfit.cluster_centers_)
labels_underfit = kmeans_underfit.labels_
segmented_underfit = centers_underfit[labels_underfit].reshape(image.shape)

# === OVERFITTING: K = 200 ===
K_overfit = 200
kmeans_overfit = KMeans(n_clusters=K_overfit, random_state=42)
kmeans_overfit.fit(pixel_vals)
centers_overfit = np.uint8(kmeans_overfit.cluster_centers_)
labels_overfit = kmeans_overfit.labels_
segmented_overfit = centers_overfit[labels_overfit].reshape(image.shape)

# === Plot Original, Underfitting, Overfitting
plt.figure(figsize=(18, 6))

plt.subplot(1, 3, 1)
plt.imshow(image)
plt.title("Original Image")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(segmented_underfit)
plt.title("Underfitting (K=2)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(segmented_overfit)
plt.title("Overfitting (K=200)")
plt.axis('off')

plt.tight_layout()
plt.show()

# === Elbow Graph: Error vs. K
K_range = list(range(2, 201, 20))  # K = 2 to 200 in steps of 20
errors = []

for K in K_range:
    kmeans = KMeans(n_clusters=K, random_state=42)
    kmeans.fit(pixel_vals)
    errors.append(kmeans.inertia_)

# === Plot Error vs. K
plt.figure(figsize=(8, 5))
plt.plot(K_range, errors, marker='o', linestyle='--')
plt.axvline(x=K_underfit, color='red', linestyle=':', label='Underfitting (K=2)')
plt.axvline(x=K_overfit, color='green', linestyle=':', label='Overfitting (K=200)')
plt.title('Error vs. Number of Clusters (Elbow Method)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Inertia (Sum of Squared Errors)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn.metrics import pairwise_distances_argmin_min

# Step 1: Generate synthetic data
n_samples = 5000
true_clusters = 5
X, _ = make_blobs(n_samples=n_samples, centers=true_clusters, cluster_std=1.0, random_state=42)

# Step 2: Add some random noise
noise_factor = 0.1
noise = noise_factor * np.random.randn(*X.shape)
X_noisy = X + noise

# Step 3: Compute error (inertia or custom distortion) for range of K
max_k = 50
errors = []

for k in range(1, max_k + 1):
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X_noisy)
    # Use distortion: sum of squared distances to closest cluster center
    _, distances = pairwise_distances_argmin_min(kmeans.cluster_centers_[kmeans.labels_], X_noisy)
    distortion = np.sum(distances ** 2)

    # Artificially increase error for very large k to simulate overfitting effect
    overfit_penalty = 0.5 * k ** 1.5 if k > true_clusters + 3 else 0
    total_error = distortion + overfit_penalty
    errors.append(total_error)

# Step 4: Plot Error vs K
plt.figure(figsize=(10, 6))
plt.plot(range(1, max_k + 1), errors, marker='o', linestyle='-')
plt.axvline(x=true_clusters, color='r', linestyle='--', label=f'Optimal K = {true_clusters}')
plt.title('Error vs Number of Clusters (K)')
plt.xlabel('Number of Clusters (K)')
plt.ylabel('Error')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt


points = {
    'A1': (2, 10), 'A2': (2, 6), 'A3': (11, 11), 'A4': (6, 9), 'A5': (6, 4),
    'A6': (1, 2), 'A7': (5, 10), 'A8': (4, 9), 'A9': (10, 12), 'A10': (7, 5),
    'A11': (9, 11), 'A12': (4, 6), 'A13': (3, 10), 'A14': (3, 8), 'A15': (6, 11)
}

data = np.array(list(points.values()))
labels = list(points.keys())


k = 3

np.random.seed(42)
initial_centroids_idx = np.random.choice(len(data), k, replace=False)
centroids = data[initial_centroids_idx]


def euclidean_distance(a, b):
    diff = a - b
    return np.sqrt(np.sum(diff ** 2))


def k_means(data, centroids, max_iters=1000, tol=1e-4):
    for _ in range(max_iters):

        clusters = [[] for _ in range(k)]


        for point in data:
            distances = [euclidean_distance(point, c) for c in centroids]
            cluster_index = np.argmin(distances)
            clusters[cluster_index].append(point)

        new_centroids = np.array([np.mean(cluster, axis=0) if cluster else centroids[i]
                                  for i, cluster in enumerate(clusters)])


        if np.all(np.abs(new_centroids - centroids) < tol):
            break

        centroids = new_centroids

    return centroids, clusters


centroids, clusters = k_means(data, centroids)


colors = ['r', 'g', 'b']
for i, cluster in enumerate(clusters):
    cluster = np.array(cluster)
    plt.scatter(cluster[:, 0], cluster[:, 1], c=colors[i], label=f'Cluster {i + 1}')

# Plot the centroids
centroids = np.array(centroids)
plt.scatter(centroids[:, 0], centroids[:, 1], c='black', marker='X', s=200, label='Centroids')

# Add labels and show plot
plt.title("K-Means Clustering (K=3)")
plt.xlabel("X")
plt.ylabel("Y")
plt.legend()
plt.grid(True)
plt.show()
```
